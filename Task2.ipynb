### Pivot Tables

```python
# Calculate average CGPA by Department and Gender

pd.pivot_table(df, values='CGPA', index='Department', columns='Gender', aggfunc='mean',)
```

```python

# Count number of students by Department and Fees Paid status
pd.pivot_table(df, values='Student_ID', index='Department', columns='Fees_Paid', aggfunc='count')
```

```python
# Calculate mean, maximum, and minimum of Midterm and Final marks by Department

pd.pivot_table(df,values=['Midterm_Marks', 'Final_Marks'], index='Department',aggfunc=['mean', 'max', 'min'])
```

```python
# Calculate total study hours per day by City, Department, and Semester

pd.pivot_table(df,values='Study_Hours_Per_Day',index=['City' , "Department"],columns="Semester" ,aggfunc='sum')
```

```python
# Count number of students with scholarships by City and Department

pd.pivot_table(df, values="Scholarship" , index="City" , columns="Department" , aggfunc="sum")
```

```python
# Calculate average CGPA and Attendance Rate by Gender and Scholarship status, including overall totals

pd.pivot_table(df,values=['CGPA', 'Attendance_Rate'],index='Gender',columns='Scholarship',aggfunc='mean',margins=True , margins_name='Total Average')
```

```python
pd.pivot_table(df , values="Final_Marks" , index="City" , columns="Department" , aggfunc="mean")
```

```python
#pivot table to calculate the average attendance rate for each city and department with semester wise

pd.pivot_table(df , values="Attendance_Rate" , index=["City","Department"] , columns="Semester" , aggfunc="mean")
```
