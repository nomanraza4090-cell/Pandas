### Data Cleaning Tasks:

```python
# Check for missing values in each column and handle them appropriately
df.isnull().sum()
```

```python
# Check for duplicate rows and remove any duplicates if found
df.duplicated().sum()
```

```python
# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# Check data types of all columns and convert if necessary
df.dtypes
```

```python
df = df.astype({'price' : float})
df.info()
```

```python
# Standardize 'region' values
df['region'] = df['region'].str.strip().str.title()
print(df['region'])
```

### Data Analysis Tasks:

```python
# Calculate total sales by region and find which region has the highest sales
df.groupby('region')['sales'].sum()
```

```python
# Find the average discount given for each product category
Discount = df.groupby('category')['discount'].mean()
Discount
```

```python
# Analyze customer ratings - calculate average rating by customer type (New vs Returning)
df.groupby('customer_type')['discount'].mean().round(2)
```

```python
# Find top 5 products by quantity sold
df.groupby('product')['quantity'].sum().sort_values(ascending=False).head(5)
```

```python
# Calculate monthly sales trend by extracting month from date column
df.groupby(df['date'].dt.to_period('M'))['sales'].sum()
```

```python
# Analyze discount impact on sales - compare average sales with and without discount
discount_impact = df.groupby(df['discount'] > 0)['sales'].agg(['mean', 'count', 'sum']).round(3)
discount = df.groupby(df['discount'] > 0)['sales'].agg
discount_impact.index = ['No Discount', 'With Discount']
print(discount_impact)
```

```python
# Find the most profitable product (highest total sales amount)
df.groupby('product')['sales'].sum().sort_values(ascending=False)
```

```python
# Analyze customer behavior - compare purchase patterns between New and Returning customers
df.groupby('customer_type')[['quantity', 'sales', 'discount']].mean().round(3)
```
